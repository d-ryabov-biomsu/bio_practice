```python
from Bio import Entrez
from Bio import SeqIO
import pandas as pd
from Bio.Align.Applications import ClustalwCommandline
from Bio import AlignIO
from tqdm.auto import tqdm


Entrez.email = 'ryaboff.dmitrij@yandex.ru'

his_table = pd.read_csv('histones.csv')
his_table.head()
```

```python
h1 = open('new_histones/H1.fasta', 'w')
h2a = open('new_histones/H2A.fasta', 'w')
h2b = open('new_histones/H2B.fasta', 'w')
h3 = open('new_histones/H3.fasta', 'w')
h4 = open('new_histones/H4.fasta', 'w')

files = [h1, h2a, h2b, h3, h4]

ref_dict = {'H1': h1, 'H2A': h2a, 'H2B': h2b, 'H3': h3, 'H4': h4}
acc_dict = {name: [] for name in ref_dict}


for i, row in tqdm(his_table.iterrows(), total=len(his_table)):
    acc = row[0]
    his_type = row[1]
    his_file = ref_dict[his_type]
    try:
        handle = Entrez.efetch(db='protein', id=acc, rettype='fasta')
    except:
        print('error', acc)
    else:
        his_file.write(handle.read())
        acc_dict[his_type].append(acc)
        
    
for file in files:
    file.close()
```
